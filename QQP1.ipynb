```
import torch
from transformers import AutoConfig, AutoModelForSequenceClassification, AutoTokenizer,AutoModelForMaskedLM
from glue.tasks import get_task, glue_tasks_num_labels
from utils.utils import *
import os 
import torch.nn as nn
import numpy as np 
import pandas as pd
```

```
ckpt = torch.load('./results_qqp/bert4-1/ckpt.pth')
print(ckpt['args'])
```

```
task_name = 'qqp'
task_repo = './glue_data/QQP'
num_labels = glue_tasks_num_labels[task_name]

config = AutoConfig.from_pretrained(
        'bert-large-uncased',
        num_labels=num_labels,
)
net = AutoModelForSequenceClassification.from_pretrained(
        'bert-large-uncased',
        config=config
)
tokenizer = AutoTokenizer.from_pretrained('bert-large-uncased')

net.cuda()

net.load_state_dict(ckpt['net'])

task = get_task(task_name, task_repo)

examples = task.get_test_examples()

label_list = task.get_labels()
label_map = {label: i for i, label in enumerate(label_list)}
print(label_map)

net_input_ids = []
net_input_masks = []
net_segment_ids = []
net_label_ids = []

for (ex_index, example) in enumerate(examples):
  net_input_id, net_input_mask, net_segment_id, net_label_id = convert_example_to_feature(example, tokenizer, 128, label_map)
  net_input_ids.append(net_input_id)
  net_input_masks.append(net_input_mask)
  net_segment_ids.append(net_segment_id)
  net_label_ids.append(net_label_id)

net_input_ids = torch.tensor(net_input_ids)
net_input_masks = torch.tensor(net_input_masks)
net_segment_ids = torch.tensor(net_segment_ids)

teset = torch.utils.data.TensorDataset(net_input_ids, net_input_masks, net_segment_ids)
prediction_sampler = torch.utils.data.SequentialSampler(teset)
teloader = torch.utils.data.DataLoader(teset, batch_size=16)
```

```
net.eval()

predictions = []
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

import torch.nn as nn
import numpy as np 

all_logits = []
# Predict 
for batch in teloader:
  # Add batch to GPU
  batch = tuple(t.to(device) for t in batch)
  # Unpack the inputs from our dataloader
  b_input_ids, b_input_mask, b_segment_ids = batch
  # Telling the model not to compute or store gradients, saving memory and speeding up prediction
  with torch.no_grad():
    # Forward pass, calculate logit predictions
    logits = net(b_input_ids, token_type_ids=b_segment_ids, attention_mask=b_input_mask)
    outputs = logits[0]

  logit = outputs.detach().cpu().numpy()
  all_logits.append(logit)
  
all_logits = np.concatenate(all_logits, axis=0)

if all_logits.shape[1] == 1:
    pred_arr = all_logits.reshape(-1)
else:
    pred_arr = np.argmax(all_logits, axis=1)

my_df = pd.DataFrame({
    "index": [str(i) for i in range(len(pred_arr))],
    "prediction":pred_arr}) 
print(my_df)

my_df.to_csv('QQP4.tsv', sep = '\t', index=False)
```
